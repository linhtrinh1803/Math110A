### 1.  

Compute the mean and volatility for Facebook using the historical data in the spreadsheet

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
fb = pd.read_csv('fb.csv').Price
fb = fb.apply(lambda x: x.replace('$',''))
price_fb = fb.to_numpy(dtype = float)
```

```python
Rn = np.zeros(shape=(19,1))
for i in range(19):
    Rn[i] = (price_fb[i+1] - price_fb[i])/price_fb[i]
```

```python
log_fb = np.log(Rn+1) 
mu_d_fb = np.sum(log_fb)/len(Rn)
mu_d_fb
```

```python
mu_hat_fb = mu_d_fb*252
mu_hat_fb
```

```python
a = (log_fb - mu_d_fb)**2
sigma_d_fb = np.sqrt(np.sum(a/18))
sigma_d_fb
```

```python
sigma_hat_fb = np.sqrt(252) *sigma_d_fb
sigma_hat_fb
```

Compute the mean and volatility for Google using the historical data in spreadsheet

```python
goog = pd.read_csv('goog.csv').Price
goog = goog.replace('\$|,','', regex=True)
price_goog = goog.to_numpy(dtype = float)
```

```python
Rn1 = np.zeros(shape=(19,1))
for i in range(19):
    Rn1[i] = (price_goog[i+1] - price_goog[i])/price_goog[i]
```

```python
log_goog = np.log(Rn1+1) 
mu_d_goog = np.sum(log_goog)/len(Rn1)
mu_d_goog
```

```python
mu_hat_goog = mu_d_goog*252
mu_hat_goog
```

```python
b = (log_goog - mu_d_goog)**2
sigma_d_goog = np.sqrt(np.sum(b/18))
sigma_d_goog
```

```python
sigma_hat_goog = np.sqrt(252) *sigma_d_goog
sigma_hat_goog
```
