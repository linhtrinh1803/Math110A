```python
import numpy as np
```

```python
price_fb = np.array([100,80,64,80,100,80,64,80,64,80,100])
```

```python
Rn = np.zeros(shape=(10,1))
for i in range(10):
    Rn[i] = (price_fb[i+1] - price_fb[i])/price_fb[i]
```

```python
log_fb = np.log(Rn+1) 
mu_d_fb = np.sum(log_fb)/10
mu_d_fb
```

```python
mu_hat_fb = mu_d_fb*252
mu_hat_fb
```

```python
a = (log_fb - mu_d_fb)**2
sigma_d_fb = np.sqrt(np.sum(a)*1/9)
sigma_d_fb
```

```python
sigma_hat_fb = np.sqrt(252) *sigma_d_fb
sigma_hat_fb
```
